```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
import plotly.graph_objs as go
import plotly.plotly as py
import scipy.stats as stats
import statsmodels.api as sm

import warnings
warnings.filterwarnings('ignore')
```

```python
weather = pd.read_csv('weatherHistory.csv')
weather.head(10)
```

```python
weather['temp_diff']=weather['Temperature (C)']- weather['Apparent Temperature (C)']
```

```python
weather.head()
```

```python
X=weather[['Humidity','Wind Speed (km/h)']]
```

```python
Y=weather['temp_diff']
```

```python
X = sm.add_constant(X)
```

```python
sonuclar = sm.OLS(Y, X).fit()
sonuclar.summary()
```

istatistiksel olarak p ye baktıgımızda constant ve degerlerimizin  anlamlı oldugunu goruyoruz.
Temp diff= -2.4381+3.0292*Humidity+0.1193*Wind Speed (km/h)
* nem oranı bir arttıgında hissedilen sıcaklık 3.0292 artmaktadır.
* ruzgar hızı bir arttıgında hissedilen sıcaklık 0.1193 artmaktadır.
ruzgar hızı arttıkca sıcaklık hıssının azalmasını beklerdım, degerlerde eksık oldugundan katsayısı yanlıs cıkmıstır.

```python
weather['Humidity_WindS']=weather['Humidity']*weather['Wind Speed (km/h)']
```

```python
X=weather[['Humidity','Wind Speed (km/h)','Humidity_WindS']]
```

```python
Y=weather['temp_diff']
```

```python
X = sm.add_constant(X)
```

```python
sonuclar = sm.OLS(Y, X).fit()
sonuclar.summary()
```

istatistiksel olarak p ye baktıgımızda constant ve degerlerimizin  anlamlı oldugunu goruyoruz.
Temp diff= -0.0839+-0.1775*Humidity+-0.0905*Wind Speed (km/h)+0.2971*Humidity_WindS
* nemdeki bir artıs, ruzgarda bir puan artanlarda artmayanlara göre 0.2971 artmaktadır
* ruzgar hızındaki bir artıs, nemde bir puan artanlarda artmayanlara göre  0.2971 artmaktadır.
